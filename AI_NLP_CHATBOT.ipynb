```python
import tensorflow as tf
from transformers import GPT2Tokenizer, TFGPT2LMHeadModel

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = TFGPT2LMHeadModel.from_pretrained("gpt2")


def generate_response(user_input, chat_history=None):
    input_ids = tokenizer.encode(user_input, return_tensors="tf")

    if chat_history is None:
        chat_history = input_ids
    else:
        chat_history = tf.concat([chat_history, input_ids], axis=1)

    response = model.generate(
        chat_history,
        max_length=100,
        pad_token_id=tokenizer.eos_token_id,
        num_return_sequences=1,
    )

    return tokenizer.decode(response[0], skip_special_tokens=True), chat_history


conversation_data = [
    ("Hello, how are you?", "I'm good, thank you!"),
    ("What are you up to?", "Just chilling and chatting."),
    ("Nice weather today, right?", "Yes, it's lovely!"),
]

chat_history = None

for user_input, expected_response in conversation_data:
    response, chat_history = generate_response(user_input, chat_history)

    print("User:", user_input)
    print("Chatbot:", response)
    print("Expected Response:", expected_response)
    print()

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Chatbot: Goodbye!")
        break
    response, chat_history = generate_response(user_input, chat_history)
    print("Chatbot:", response)
```
